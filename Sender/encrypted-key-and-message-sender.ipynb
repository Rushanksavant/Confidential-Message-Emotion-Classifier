# Defining important Functions:

```python
def generate_keys():
    '''This returns pair of Private and Public keys'''
    length= 256*25
    private_key= RSA.generate(length, Random.new().read)
    public_key= private_key.publickey()
    return private_key, public_key


def encrypt(public_key, mssg):
    """
    This fnction encrypts the message(mssg)
    using the public key
    """
    return base64.b64encode(public_key.encrypt(mssg.encode(), 0)[0])
```
